```python
class Node():

    def __init__(self, element, parentNode):
        self.element = element
        self.parent = parentNode
        self.right = None
        self.left = None
```

```python
class BinarySearchTree():

    def __init__(self):
        print 'BinarySearchTree invoked.'
        self.root = None
        self.size = 0

    def add(self, element):
        print 'Add Invoked with element', element
        if self.root is None:
            if element is None:
                raise TypeError
            self.root = Node(element, None)
            self.size += 1
            print 'root created.'
            return True
        else:
            temp = self.root
            while True:
                if element == temp.element:
                    print 'Duplicate element'
                    return False
                elif element < temp.element:
                    if temp.left is not None:
                        temp = temp.left
                    else:
                        temp.left = Node(element, temp)
                        self.size += 1
                        print 'Left Node created.'
                        return True
                elif temp.right is not None:
                    temp = temp.right
                else:
                    temp.right = Node(element, temp)
                    self.size += 1
                    print 'Right Node created.'
                    return True
            print 'Added', element


    def successor(self, e):
        if e is None:
            return None
        elif e.right is not None:
            #successor is leftmost Entry in right subtree of e
            p = e.right
            while p.left is not None:
                p = p.left
            return p
        else:
            # go up the tree to the left as far as possible, then go up
            # to the right.
            p = e.parent
            ch = e
            while p is not None and ch == p.right:
                ch = p
                p = p.parent
            return p

    def tree_size(self):
        return self.size

    def getNode(self, obj):
        if obj is None:
            raise TypeError
        e = self.root
        while e is not None: 
            if obj == e.element:
                return e
            elif obj < e.element:
                e = e.left
            else:
                e = e.right
        return None

    def deleteNode(self, p):
        self.size -= 1

        #If p has two children, replace p's element with p's successor's
        #element, then make p reference that successor.
        if p.left is not None and p.right is not None:
            s = successor (p)
            p.element = s.element
            p = s

        #At this point, p has either no children or one child.
        if p.left is not None:
            replacement = p.left
        else:
            replacement = p.right

        #If p has at least one child, link replacement to p.parent.
        if replacement is not None:
            replacement.parent = p.parent
            if p.parent is None:
                root = replacement
            elif p == p.parent.left:
                p.parent.left  = replacement
            else:
                p.parent.right = replacement
        elif p.parent is None:
            root = None
        else:
            if p == p.parent.left:
                p.parent.left = None
            else:
                p.parent.right = None
        return p

    def remove(self, obj):
        e = self.getNode(obj)
        if e is None:
            return False
        self.deleteNode (e)
        return True

    def contains(self,obj):
        print obj
        return self.getNode (obj) is not None
    
    def equals(self, obj):
        if not isinstance(obj, BinaryTree):
            return False
        return equals (self.root, obj.root)

    def equals(self, nodeA, nodeB):
        if nodeA is None or nodeB is None:
            return nodeA == nodeB
        if nodeA.element != nodeB.element:
            return False
        if self.equals(nodeA.left, nodeB.left) and self.equals(nodeA.right, nodeB.right):
            return True
        return False

    def copy(self, sourceNode ,parentNode):
        if sourceNode != None:
            targetNode = Node(sourceNode.element, parentNode)
            sourceNode.left = self.copy(sourceNode.left, targetNode)
            sourceNode.right = self.copy(sourceNode.right, targetNode)
            return targetNode
        return None
```
